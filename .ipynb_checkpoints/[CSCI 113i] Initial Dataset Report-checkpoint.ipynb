```python
#Completeness
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dynamic_pricing.csv')

'''
Using the .isnull() and .isna() of the Pandas library, we check for None or NaN  values in the dataset, and use the .sum() to sum the values. 
As we can see there are no null values, thus all values in the dataset are present and no further imputation or deletion is needed
'''

data_null = df.isnull().sum()
data_na = df.isna().sum()
print(f"Summary of Data Nulls \n \n{data_null}\n")
print(f"Summary of Data NA \n \n{data_na}")
```

```python
#Completeness and Consistency
 #Checking the datatypes of the columns of the dataset and the number of non-null values "
print(df.info())
```
